```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# minus 폰트 깨짐 방지
plt.rc('axes', unicode_minus = False)

#한글 폰트 설정
plt.rc('font', family='NanumGothic')

# retina 설정
from IPython.display import set_matplotlib_formats

set_matplotlib_formats('retina')

# csv 파일 불러오기
df = pd.read_csv("data file/소상공인시장진흥공단_상가업소정보_의료기관_201909.csv", low_memory = False)
```

```python
df_seoul = df[df["시도명"] == "서울특별시"]
df_seoul.shape
```

```python
df_seoul["시군구명"].value_counts()
```

서울시의 데이터를 샘플링하고 서울시의 구별로 병원의 수를 알아보았습니다.

```python
df_seoul ["시군구명"].value_counts().plot.bar(figsize=(10, 4), rot=30)
```

bar plot를 그리고 사이즈와 글자의 회전각도도 변경했습니다.

```python
plt.figure(figsize=(15, 4))
sns.countplot(data=df_seoul, x="시군구명")
```

seaborn의 countplot를 그려봅니다.

```python
df_seoul[["경도", "위도", "시군구명"]].plot.scatter(x="경도", y="위도", figsize=(8, 7), grid=True)
```

```python
plt.figure(figsize=(9, 8))
sns.scatterplot(data=df_seoul, x="경도", y="위도", hue="상권업종중분류명")
```

```python
plt.figure(figsize=(16, 12))
sns.scatterplot(data=df_seoul, x="경도", y="위도", hue="시군구명")
```

scatter plot을 matplotlib를 그려보고 hue를 이용하여 상업분류별과 지역별로 그려봅니다
