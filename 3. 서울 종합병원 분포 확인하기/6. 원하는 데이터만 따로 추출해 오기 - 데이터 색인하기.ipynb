## 2개의 조건 사용하기

```python
df[df["상권업종대분류명"] == "의료"]
```

```python
df.loc[df["상권업종대분류명"] == "의료"]["상권업종중분류명"]
```

상권업종대분류명이 의료인 데이터의 상권업종중분류명을 가져오고 싶으면 **loc**를 함께 사용합니다. 하지만 위의 방법은 의료를 검색하고, 또 다음조건을 검색하므로 속도가 느립니다. 그래서 다음 방법을 사용합니다.

```python
df.loc[df["상권업종대분류명"] == "의료", "상권업종중분류명"]
```

## 조건에 맞는 데이터 개수 세기

```python
df.loc[df["상권업종대분류명"] == "의료", "상권업종중분류명"].value_counts()
```

value_counts()로 데티터의 갯수를 셀 수 있습니다.

```python
m = df["상권업종대분류명"] == "의료"
df.loc[m, "상권업종중분류명"].value_counts()
```

boolean indexing으로 위와 같이 코드를 작성할 수도 있습니다.
