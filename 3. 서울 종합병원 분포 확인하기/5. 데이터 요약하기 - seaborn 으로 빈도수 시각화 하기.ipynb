```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# minus 폰트 깨짐 방지
plt.rc('axes', unicode_minus = False)

#한글 폰트 설정
plt.rc('font', family='NanumGothic')

# retina 설정
from IPython.display import set_matplotlib_formats

set_matplotlib_formats('retina')

# csv 파일 불러오기
df = pd.read_csv("data file/소상공인시장진흥공단_상가업소정보_의료기관_201909.csv", low_memory = False)
```

## seaborn으로 그래프 그려보기

```python
sns.countplot(data = df, y ="시도명")
```

seaborn은 x, y, data를 기본으로 넣어줘야 기본으로 넣어줘야 하지만 countplot은 x,y 중 하나만 넣어도 됩니다. <br/> seaborn 장점 : 고급 통계 기능을 그래프 내부에서 제공합니다. <br/> seaborn 단점 : 데이테 크기가 클수로 속도가 느립니다.

## 그래프로 그리는 데이터 가공하기

```python
df["상권업종대분류명"].value_counts()
```

```python
c = df["상권업종중분류명"].value_counts()
c
```

```python
n = df["상권업종중분류명"].value_counts(normalize=True)
n
```

## 그래프 그리기

```python
c.plot.bar(rot=1)
```

```python
#파이그래프
n.plot.pie()
```

```python
#bar 그래프
c = df["상권업종소분류명"].value_counts()
c.plot.barh()
```
