## 그룹화된 요약값 보기

```python
df["시도명"].value_counts()
```

value_counts()는 각 value가 몇 개인지를 세워줍니다.

```python
df["시도명"].value_counts(normalize = True)
```

normalize 옵션을 넣으면 전체에서 차지하는 비율을 계산해줍니다.

```python
city_normalize = df["시도명"].value_counts(normalize = True)
city_normalize.plot.barh()
```

위의 비율 값을 city_noralize 변수에 담아 그래프로 나타 냈습니다.

```python
city_normalize.plot.pie(figsize=(7, 7))
```

파이 차트를 그려보고 figsize로 사이즈를 조절해주었습니다.
* 전체 비율을 한눈에 볼 때에는 파이차트가 편합니다.
* 차이를 확인하기에는 막대그래프가 편합니다.
